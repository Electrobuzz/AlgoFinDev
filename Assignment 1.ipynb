```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define the ticker symbol for ITC
ticker = "ITC.NS"

# Retrieve historical data from Yahoo Finance
data = yf.download(ticker, period="3mo")

# Calculate the Simple Moving Average (SMA)
sma = data['Close'].rolling(window=5).mean()

# Calculate the Exponential Moving Average (EMA)
ema = data['Close'].ewm(span=5, adjust=False).mean()

# Plotting the data
plt.figure(figsize=(10, 6))

#Plotting Simple Moving Average(SMA)
plt.plot(data.index, sma, label='Simple Moving Average (SMA)')

#plotting Exponential Moving Average(EMA)
plt.plot(data.index, ema, label='Exponential Moving Average (EMA)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('ITC Stock Moving Averages')
plt.legend()

plt.show()
```
